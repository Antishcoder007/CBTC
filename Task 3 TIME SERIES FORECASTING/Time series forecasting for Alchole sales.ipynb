```python
## perform Dickey-Fuller test

from statsmodels.tsa.stattools import adfuller
dftest = adfuller(Alchole_Sales['Sales'])
```

```python
#Ho: it is non staionary
#H1: it is stationary
def adfull_test(Sales):
    result = adfuller(Sales)
    labels =['ADF Test Statistic','p-value','#lags Used','Number of observation Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
    if result[1] <= 0.05:
        print("stationary")
    else:
        print("non-Stationary")
```

```python
adfull_test(Alchole_Sales['Sales'])
```

```python
Alchole_Sales['Sales First Difference'] = Alchole_Sales['Sales'] - Alchole_Sales['Sales'].shift(1)
```

```python
Alchole_Sales['Seasonal First Difference'] = Alchole_Sales['Sales'] - Alchole_Sales['Sales'].shift(12)
```

```python
## again test dickey fuller test
adfull_test(Alchole_Sales['Seasonal First Difference'].dropna())
```

```python
# form non-seasional data
import statsmodels.api as sm
```

```python
model = sm.tsa.statespace.SARIMAX(Alchole_Sales['Sales'],order=(1,1,1),seasonal_order=(1,1,1,12))
result = model.fit()
```
